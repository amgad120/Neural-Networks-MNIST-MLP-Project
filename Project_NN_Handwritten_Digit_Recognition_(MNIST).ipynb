```python
# =========================================================
# Neural Networks Project
# MNIST Handwritten Digit Recognition using MLP
# =========================================================

# =========================================================
# Import Libraries
# =========================================================

import numpy as np
import matplotlib.pyplot as plt

from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical

from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

# =========================================================
# Load Dataset
# =========================================================

(x_train, y_train), (x_test, y_test) = mnist.load_data()

print("Training Data Shape:", x_train.shape)
print("Testing Data Shape:", x_test.shape)

# =========================================================
# Display Sample Images
# =========================================================

plt.figure(figsize=(10,5))

for i in range(10):
    plt.subplot(2,5,i+1)
    plt.imshow(x_train[i], cmap='gray')
    plt.title(f"Label: {y_train[i]}")
    plt.axis('off')

plt.tight_layout()
plt.show()

# =========================================================
# Data Preprocessing
# =========================================================

# Normalize Data
x_train = x_train / 255.0
x_test = x_test / 255.0

# Reshape Images
x_train = x_train.reshape(60000, 784)
x_test = x_test.reshape(10000, 784)

print("New Training Shape:", x_train.shape)
print("New Testing Shape:", x_test.shape)

# One Hot Encoding
y_train_encoded = to_categorical(y_train)
y_test_encoded = to_categorical(y_test)

print("Encoded Label Example:")
print(y_train_encoded[0])

# =========================================================
# Experiment 1 - ReLU Activation Function
# =========================================================

model_relu = Sequential()

# Input + Hidden Layer 1
model_relu.add(Dense(
    128,
    activation='relu',
    input_shape=(784,)
))

# Hidden Layer 2
model_relu.add(Dense(
    64,
    activation='relu'
))

# Dropout Layer
model_relu.add(Dropout(0.3))

# Output Layer
model_relu.add(Dense(
    10,
    activation='softmax'
))

# =========================================================
# Compile Model
# =========================================================

model_relu.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# =========================================================
# Model Summary
# =========================================================

model_relu.summary()

# =========================================================
# Train Model
# =========================================================

history_relu = model_relu.fit(
    x_train,
    y_train_encoded,
    epochs=20,
    batch_size=128,
    validation_split=0.2
)

# =========================================================
# Evaluate Model
# =========================================================

test_loss_relu, test_accuracy_relu = model_relu.evaluate(
    x_test,
    y_test_encoded
)

print("\n========== Experiment 1 Results (ReLU) ==========")
print(f"Test Loss: {test_loss_relu:.4f}")
print(f"Test Accuracy: {test_accuracy_relu:.4f}")

# =========================================================
# Predictions
# =========================================================

predictions_relu = model_relu.predict(x_test)

predicted_classes_relu = np.argmax(predictions_relu, axis=1)

# =========================================================
# Display Predictions
# =========================================================

plt.figure(figsize=(12,6))

for i in range(10):
    plt.subplot(2,5,i+1)
    plt.imshow(x_test[i].reshape(28,28), cmap='gray')
    plt.title(f"Predicted: {predicted_classes_relu[i]}")
    plt.axis('off')

plt.tight_layout()
plt.show()

# =========================================================
# Classification Report
# =========================================================

print("\n========== Classification Report ==========")

print(classification_report(
    y_test,
    predicted_classes_relu
))
# =========================================================
# Confusion Matrix
# =========================================================

cm = confusion_matrix(
    y_test,
    predicted_classes_relu
)

plt.figure(figsize=(10,8))

plt.imshow(cm, cmap='Blues')

plt.title("Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")

plt.colorbar()

classes = np.arange(10)

plt.xticks(classes)
plt.yticks(classes)

for i in range(10):
    for j in range(10):
        plt.text(
            j,
            i,
            cm[i, j],
            ha='center',
            va='center',
            color='black'
        )

plt.show()
# =========================================================
# Plot Loss Curve
# =========================================================

plt.figure(figsize=(8,5))

plt.plot(history_relu.history['loss'])
plt.plot(history_relu.history['val_loss'])

plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')

plt.legend(['Train', 'Validation'])

plt.show()

# =========================================================
# Plot Accuracy Curve
# =========================================================

plt.figure(figsize=(8,5))

plt.plot(history_relu.history['accuracy'])
plt.plot(history_relu.history['val_accuracy'])

plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')

plt.legend(['Train', 'Validation'])

plt.show()

# =========================================================
# =========================================================
# Experiment 2 - Tanh Activation Function
# =========================================================
# =========================================================

model_tanh = Sequential()

# Input + Hidden Layer 1
model_tanh.add(Dense(
    128,
    activation='tanh',
    input_shape=(784,)
))

# Hidden Layer 2
model_tanh.add(Dense(
    64,
    activation='tanh'
))

# Dropout Layer
model_tanh.add(Dropout(0.3))

# Output Layer
model_tanh.add(Dense(
    10,
    activation='softmax'
))

# =========================================================
# Compile Second Model
# =========================================================

model_tanh.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# =========================================================
# Train Second Model
# =========================================================

history_tanh = model_tanh.fit(
    x_train,
    y_train_encoded,
    epochs=20,
    batch_size=128,
    validation_split=0.2
)

# =========================================================
# Evaluate Second Model
# =========================================================

test_loss_tanh, test_accuracy_tanh = model_tanh.evaluate(
    x_test,
    y_test_encoded
)

print("\n========== Experiment 2 Results (Tanh) ==========")
print(f"Test Loss: {test_loss_tanh:.4f}")
print(f"Test Accuracy: {test_accuracy_tanh:.4f}")

# =========================================================
# Compare Experiments
# =========================================================

print("\n=================================================")
print("Comparison Between Experiments")
print("=================================================")

print(f"""
Experiment 1:
Activation Function = ReLU
Accuracy = {test_accuracy_relu:.4f}
Loss = {test_loss_relu:.4f}

Experiment 2:
Activation Function = Tanh
Accuracy = {test_accuracy_tanh:.4f}
Loss = {test_loss_tanh:.4f}
""")

# =========================================================
# Accuracy Comparison Curve
# =========================================================

plt.figure(figsize=(8,5))

plt.plot(history_relu.history['val_accuracy'])
plt.plot(history_tanh.history['val_accuracy'])

plt.title('Validation Accuracy Comparison')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')

plt.legend(['ReLU', 'Tanh'])

plt.show()

# =========================================================
# Loss Comparison Curve
# =========================================================

plt.figure(figsize=(8,5))

plt.plot(history_relu.history['val_loss'])
plt.plot(history_tanh.history['val_loss'])

plt.title('Validation Loss Comparison')
plt.xlabel('Epoch')
plt.ylabel('Loss')

plt.legend(['ReLU', 'Tanh'])

plt.show()

# =========================================================
# Final Results Table
# =========================================================

print("\n=================================================")
print("Final Results Table")
print("=================================================")

print("Experiment\tActivation\tAccuracy\tLoss")

print(f"Exp 1\t\tReLU\t\t{test_accuracy_relu:.4f}\t\t{test_loss_relu:.4f}")

print(f"Exp 2\t\tTanh\t\t{test_accuracy_tanh:.4f}\t\t{test_loss_tanh:.4f}")

# =========================================================
# End of Project
# =========================================================
```
