```python
# =========================================================
# Neural Networks Project
# MNIST Handwritten Digit Recognition using MLP
# Full Professional Version
# =========================================================

# =========================================================
# Import Libraries
# =========================================================

import os
import numpy as np
import matplotlib.pyplot as plt

from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Input
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping

from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

# =========================================================
# Create Project Folders
# =========================================================

os.makedirs("images", exist_ok=True)
os.makedirs("results", exist_ok=True)
os.makedirs("model", exist_ok=True)

# =========================================================
# Load Dataset
# =========================================================

(x_train, y_train), (x_test, y_test) = mnist.load_data()

print("Training Data Shape:", x_train.shape)
print("Testing Data Shape:", x_test.shape)

# =========================================================
# Display Sample Images
# =========================================================

plt.figure(figsize=(10,5))

for i in range(10):

    plt.subplot(2,5,i+1)

    plt.imshow(x_train[i], cmap='gray')

    plt.title(f"Label: {y_train[i]}")

    plt.axis('off')

plt.tight_layout()

plt.savefig("images/sample_digits.png")

plt.show()

# =========================================================
# Data Preprocessing
# =========================================================

# Normalize Data
x_train = x_train / 255.0
x_test = x_test / 255.0

# Reshape Images
x_train = x_train.reshape(60000, 784)
x_test = x_test.reshape(10000, 784)

print("New Training Shape:", x_train.shape)
print("New Testing Shape:", x_test.shape)

# One Hot Encoding
y_train_encoded = to_categorical(y_train)
y_test_encoded = to_categorical(y_test)

print("Encoded Label Example:")
print(y_train_encoded[0])

# =========================================================
# Build Model Function
# =========================================================

def build_model(
    activation='relu',
    neurons=128
):

    model = Sequential([

        Input(shape=(784,)),

        Dense(
            neurons,
            activation=activation
        ),

        Dense(
            64,
            activation=activation
        ),

        Dropout(0.3),

        Dense(
            10,
            activation='softmax'
        )

    ])

    model.compile(
        optimizer='adam',
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    return model

# =========================================================
# Experiment 1
# ReLU + 128 Neurons
# =========================================================

print("\n=================================================")
print("Experiment 1")
print("ReLU + 128 Neurons")
print("=================================================")

model_relu = build_model(
    activation='relu',
    neurons=128
)

model_relu.summary()

history_relu = model_relu.fit(
    x_train,
    y_train_encoded,
    epochs=20,
    batch_size=128,
    validation_split=0.2,
    verbose=1
)

# Evaluate Model
test_loss_relu, test_accuracy_relu = model_relu.evaluate(
    x_test,
    y_test_encoded
)

print(f"\nExperiment 1 Loss: {test_loss_relu:.4f}")
print(f"Experiment 1 Accuracy: {test_accuracy_relu:.4f}")

# =========================================================
# Experiment 2
# Tanh + 128 Neurons
# =========================================================

print("\n=================================================")
print("Experiment 2")
print("Tanh + 128 Neurons")
print("=================================================")

model_tanh = build_model(
    activation='tanh',
    neurons=128
)

history_tanh = model_tanh.fit(
    x_train,
    y_train_encoded,
    epochs=20,
    batch_size=128,
    validation_split=0.2,
    verbose=1
)

# Evaluate Model
test_loss_tanh, test_accuracy_tanh = model_tanh.evaluate(
    x_test,
    y_test_encoded
)

print(f"\nExperiment 2 Loss: {test_loss_tanh:.4f}")
print(f"Experiment 2 Accuracy: {test_accuracy_tanh:.4f}")

# =========================================================
# Experiment 3
# ReLU + 256 Neurons
# =========================================================

print("\n=================================================")
print("Experiment 3")
print("ReLU + 256 Neurons")
print("=================================================")

model_exp3 = build_model(
    activation='relu',
    neurons=256
)

history_exp3 = model_exp3.fit(
    x_train,
    y_train_encoded,
    epochs=20,
    batch_size=128,
    validation_split=0.2,
    verbose=1
)

# Evaluate Model
test_loss_exp3, test_accuracy_exp3 = model_exp3.evaluate(
    x_test,
    y_test_encoded
)

print(f"\nExperiment 3 Loss: {test_loss_exp3:.4f}")
print(f"Experiment 3 Accuracy: {test_accuracy_exp3:.4f}")

# =========================================================
# Compare Experiments
# =========================================================

print("\n=================================================")
print("Final Results Table")
print("=================================================")

print("Experiment\tActivation\tNeurons\tAccuracy\tLoss")

print(
    f"Exp 1\t\tReLU\t\t128\t\t{test_accuracy_relu:.4f}\t\t{test_loss_relu:.4f}"
)

print(
    f"Exp 2\t\tTanh\t\t128\t\t{test_accuracy_tanh:.4f}\t\t{test_loss_tanh:.4f}"
)

print(
    f"Exp 3\t\tReLU\t\t256\t\t{test_accuracy_exp3:.4f}\t\t{test_loss_exp3:.4f}"
)

# =========================================================
# Save Results
# =========================================================

with open("results/final_results.txt", "w") as file:

    file.write("Final Results Table\n")
    file.write("=================================================\n")

    file.write(
        f"Experiment 1 | ReLU | 128 Neurons | Accuracy: {test_accuracy_relu:.4f} | Loss: {test_loss_relu:.4f}\n"
    )

    file.write(
        f"Experiment 2 | Tanh | 128 Neurons | Accuracy: {test_accuracy_tanh:.4f} | Loss: {test_loss_tanh:.4f}\n"
    )

    file.write(
        f"Experiment 3 | ReLU | 256 Neurons | Accuracy: {test_accuracy_exp3:.4f} | Loss: {test_loss_exp3:.4f}\n"
    )

# =========================================================
# Comparison Accuracy Curve
# =========================================================

plt.figure(figsize=(8,5))

plt.plot(history_relu.history['val_accuracy'])

plt.plot(history_tanh.history['val_accuracy'])

plt.plot(history_exp3.history['val_accuracy'])

plt.title('Validation Accuracy Comparison')

plt.xlabel('Epoch')

plt.ylabel('Accuracy')

plt.legend([
    'ReLU',
    'Tanh',
    'ReLU + 256'
])

plt.savefig(
    "images/validation_accuracy_comparison.png"
)

plt.show()

# =========================================================
# Comparison Loss Curve
# =========================================================

plt.figure(figsize=(8,5))

plt.plot(history_relu.history['val_loss'])

plt.plot(history_tanh.history['val_loss'])

plt.plot(history_exp3.history['val_loss'])

plt.title('Validation Loss Comparison')

plt.xlabel('Epoch')

plt.ylabel('Loss')

plt.legend([
    'ReLU',
    'Tanh',
    'ReLU + 256'
])

plt.savefig(
    "images/validation_loss_comparison.png"
)

plt.show()

# =========================================================
# =========================================================
# BEST FINAL MODEL
# ReLU + 256 + EarlyStopping
# =========================================================
# =========================================================

print("\n=================================================")
print("BEST FINAL MODEL")
print("ReLU + 256 + EarlyStopping")
print("=================================================")

# =========================================================
# EarlyStopping
# =========================================================

early_stop = EarlyStopping(
    monitor='val_loss',
    patience=3,
    restore_best_weights=True
)

# =========================================================
# Build Final Model
# =========================================================

model_best = Sequential([

    Input(shape=(784,)),

    Dense(
        256,
        activation='relu'
    ),

    Dense(
        64,
        activation='relu'
    ),

    Dropout(0.3),

    Dense(
        10,
        activation='softmax'
    )

])

# =========================================================
# Compile Model
# =========================================================

model_best.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# =========================================================
# Model Summary
# =========================================================

model_best.summary()

# =========================================================
# Train Final Model
# =========================================================

history_best = model_best.fit(
    x_train,
    y_train_encoded,
    epochs=20,
    batch_size=128,
    validation_split=0.2,
    callbacks=[early_stop],
    verbose=1
)

# =========================================================
# Evaluate Final Model
# =========================================================

test_loss_best, test_accuracy_best = model_best.evaluate(
    x_test,
    y_test_encoded
)

print("\n=================================================")
print("BEST MODEL RESULTS")
print("=================================================")

print(f"Test Loss: {test_loss_best:.4f}")

print(f"Test Accuracy: {test_accuracy_best:.4f}")

# =========================================================
# Predictions
# =========================================================

predictions_best = model_best.predict(x_test)

predicted_classes_best = np.argmax(
    predictions_best,
    axis=1
)

# =========================================================
# Show Sample Predictions
# =========================================================

plt.figure(figsize=(12,6))

for i in range(10):

    plt.subplot(2,5,i+1)

    plt.imshow(
        x_test[i].reshape(28,28),
        cmap='gray'
    )

    plt.title(
        f"Predicted: {predicted_classes_best[i]}"
    )

    plt.axis('off')

plt.tight_layout()

plt.savefig("images/sample_predictions.png")

plt.show()

# =========================================================
# Classification Report
# =========================================================

print("\n========== Classification Report ==========")

report_best = classification_report(
    y_test,
    predicted_classes_best
)

print(report_best)

# Save Classification Report
with open(
    "results/classification_report.txt",
    "w"
) as file:

    file.write(report_best)

# =========================================================
# Confusion Matrix
# =========================================================

cm_best = confusion_matrix(
    y_test,
    predicted_classes_best
)

plt.figure(figsize=(10,8))

plt.imshow(
    cm_best,
    cmap='Blues'
)

plt.title(
    "Best Model Confusion Matrix"
)

plt.xlabel("Predicted Label")

plt.ylabel("True Label")

plt.colorbar()

classes = np.arange(10)

plt.xticks(classes)

plt.yticks(classes)

for i in range(10):

    for j in range(10):

        plt.text(
            j,
            i,
            cm_best[i, j],
            ha='center',
            va='center',
            color='white'
            if cm_best[i, j] > 500
            else 'black'
        )

plt.savefig(
    "images/confusion_matrix.png"
)

plt.show()

# =========================================================
# Final Model Accuracy Curve
# =========================================================

plt.figure(figsize=(8,5))

plt.plot(
    history_best.history['accuracy']
)

plt.plot(
    history_best.history['val_accuracy']
)

plt.title(
    'Best Model Accuracy'
)

plt.xlabel('Epoch')

plt.ylabel('Accuracy')

plt.legend([
    'Train',
    'Validation'
])

plt.savefig(
    "images/accuracy_curve.png"
)

plt.show()

# =========================================================
# Final Model Loss Curve
# =========================================================

plt.figure(figsize=(8,5))

plt.plot(
    history_best.history['loss']
)

plt.plot(
    history_best.history['val_loss']
)

plt.title(
    'Best Model Loss'
)

plt.xlabel('Epoch')

plt.ylabel('Loss')

plt.legend([
    'Train',
    'Validation'
])

plt.savefig(
    "images/loss_curve.png"
)

plt.show()

# =========================================================
# Save Final Model
# =========================================================

model_best.save(
    "model/best_mnist_model.keras"
)

print("\nBest Model Saved Successfully!")

# =========================================================
# PROJECT FINISHED
# =========================================================

print("\n=================================================")
print("PROJECT COMPLETED SUCCESSFULLY")
print("=================================================")
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import load_model
from PIL import Image
import matplotlib.pyplot as plt

# =========================
# Load trained model
# =========================
model = load_model("model/best_mnist_model.keras")

# =========================
# Function to preprocess image
# =========================
def preprocess_image(image_path):
    # Open image in grayscale
    img = Image.open(image_path).convert('L')

    # Resize to 28x28 (MNIST format)
    img = img.resize((28, 28))

    # Convert to numpy array
    img_array = np.array(img)

    # Normalize (0-1)
    img_array = img_array / 255.0

    # Flatten (784,)
    img_array = img_array.reshape(1, 784)

    return img_array, img

# =========================
# Prediction function
# =========================
def predict_digit(image_path):
    img_array, img = preprocess_image(image_path)

    prediction = model.predict(img_array)
    predicted_class = np.argmax(prediction)

    # Show image
    plt.imshow(img, cmap='gray')
    plt.title(f"Predicted Digit: {predicted_class}")
    plt.axis('off')
    plt.show()

    return predicted_class

# =========================
# TEST IMAGE
# =========================
image_path = "/content/sample_image-300x298.webp"  # <-- حط صورة هنا

result = predict_digit(image_path)
print("Predicted Digit:", result)
```
